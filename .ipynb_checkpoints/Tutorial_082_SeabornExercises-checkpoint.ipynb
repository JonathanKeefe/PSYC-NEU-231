# Seaborn Exercises

1) From the "tips" dataset
    1. add a percentage tip column to the dataset
    2. visualize the relationship between tip percentage, with day of week and time 
    3. visualize tip ~ day, time, and number of people (which function allows three variables independent variables at once?)


```python
import seaborn as sns
import matplotlib.pyplot as plt # not strictly necessary for most plotting applications
import pandas as pd
```

```python
#  First we will load the Iris flower dataset, this dataset consists of severa, measurements made from 50 flowers 
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Now let's look at the associations between the variables
sns.pairplot(data=tips,hue='size')
```

a) Percentage tip column

```python
percentage_tip = pd.DataFrame(tips.tip/tips.total_bill)
percentage_tip.columns = ['percentage']
full_tips = pd.concat([tips, percentage_tip], axis = 1)
print(full_tips)
```

b) Visualize the relationship between tip percentage, with day of week and time

```python
sns.pointplot(x='day',y='percentage',hue='time',data=full_tips)
```

c) visualize tip ~ day, time, and number of people (which function allows three variables independent variables at once?)

```python
sns.pairplot(data=full_tips,hue='size')

sns.catplot(x='size',y='tip',hue='day',col='time',data=full_tips,kind='point')
plt.show()
```

2) Find an interesting trend in the "exercise" dataset

```python
exercise = sns.load_dataset('exercise')
exercise.head()
```

```python
sns.pointplot(x = 'time', y = 'pulse', hue = 'kind', data=exercise)
```

The plot above demonstrates that pulse only increases with increasing exercise duration when participants are running - not walking or running.
